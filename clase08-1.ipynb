```python
cols=df.columns
cols=cols.drop("calificaciones")
x_list = []

for col in cols:
    x_list.append(df[col].values.reshape(-1,1))

n=len(df)                                      
ones=np.ones([n,1])
x=np.hstack(x_list)
X = np.hstack((ones,x))
y=df['calificaciones']
ols=sm.OLS(y,X)
results=ols.fit()
results.summary()
```
