```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
df=pd.read_excel(r"D:\Motor Trend Car Road Tests.xlsx")
df.describe()
```

```python
df.head(15)
```

```python
df2=df.drop_duplicates()
df2.describe()
df2=df2.drop(columns=['model'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
cols=df2.columns
cols=cols.drop('mpg')

x_list = []

for col in cols:
    x_list.append(df2[col].values.reshape(-1,1))

n=len(df2)                                      
ones=np.ones([n,1])
x=np.hstack(x_list)
X = np.hstack((ones,x))
y=df['mpg']
xtrain,xtest,ytrain,ytest=train_test_split(X,y,train_size=0.60,random_state=137)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(xtrain)
Xscaled_train=scaler.fit_transform(xtrain)
ones=np.ones((len(xtrain),1))
X_e_train=np.hstack((ones,Xscaled_train))
modelo=sm.OLS(ytrain,X_e_train)
results=modelo.fit()
results.summary()
```

```python
#Ahora probamos el modelo con los datos test
x_test_scaled = sm.add_constant(Xscaled_train)  # Agregar constante
y_pred = results.predict(x_test_scaled)
LinearRegression(ytest,y_pred)
LinearRegression.score(ytest,y_pred)
```
