```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

```python
df=pd.read_excel(r"D:\Motor Trend Car Road Tests (1).xlsx")
```

```python
df.describe()
```

**1.1**

```python
df2=df.drop(columns='model')
df2.columns
```

```python
cols=df2.columns
cols=cols.drop('mpg')

x_list = []

for col in cols:
    x_list.append(df2[col].values.reshape(-1,1))

n=len(df2)                                      
ones=np.ones([n,1])
x=np.hstack(x_list)
x=np.hstack([ones,x])
y=df2['mpg']
modelo=sm.OLS(y,x)
results=modelo.fit()
results.summary()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.40,random_state=137)
scaler=StandardScaler()
Xscaled_train=scaler.fit_transform(xtrain)
ones=np.ones((len(xtrain),1))
X_e_train=np.hstack((ones,Xscaled_train))
modelo=sm.OLS(ytrain,X_e_train)
results=modelo.fit()
results.summary()
```

Aca los coeficientes de son enormes, los mas grandes son x2, x9 y x10

```python
Xscaled_test = scaler.transform(xtest)
x_test_scaled = sm.add_constant(Xscaled_test)
y_pred_test = results.predict(x_test_scaled)
r2_test = r2_score(ytest, y_pred_test)
print(f"R^2 Train: 0.990")
print(f"R^2 Test: {r2_test}")
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge=Ridge(alpha=1)
ridge.fit(x,y)
print(f"Coeficientes:{ ridge.coef_}Intercepto:{ridge.intercept_} R^2:{ridge.score(x,y)}")
```

```python
dfd=pd.get_dummies(df2,columns=['cyl','gear','carb'])

dfd.head(32)
```

```python
cols=dfd.columns
cols=cols.drop('mpg')
x_list = []

for col in cols:
    x_list.append(dfd[col].values.reshape(-1,1))

n=len(dfd)                                      
ones=np.ones([n,1])
x=np.hstack(x_list)
x=np.hstack([ones,x])
x = np.asarray(x, dtype=float)  
y=dfd['mpg']
modelo=sm.OLS(y,x)
results=modelo.fit()
results.summary()
```

la variable mas importante es la ultima pero tenemos una constante enorme 

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size=0.40, random_state=137)
scaler = StandardScaler().fit(xtrain)
Xscaled_train = scaler.transform(xtrain)
Xscaled_test = scaler.transform(xtest)
X_e_train = sm.add_constant(Xscaled_train) 
X_e_test = sm.add_constant(Xscaled_test)
modelo = sm.OLS(ytrain, X_e_train)
results = modelo.fit()
y_pred_train = results.predict(X_e_train)
r2_train = r2_score(ytrain, y_pred_train)
r2_test = r2_score(ytest, y_pred_test)
print(f"R Train: {r2_train}")
print(f"R Test: {r2_test}")
```

no hay mucho que decir, el overfit esta brutal
