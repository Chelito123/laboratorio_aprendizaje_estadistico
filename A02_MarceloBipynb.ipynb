```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
datos=pd.read_excel(r'D:\Boston Housing Dataset 1978.xlsx')
```

```python
datos.describe()
```

```python
datos2=datos.dropna()
datos2.describe()
```

```python
y=datos2['MEDV'].values.reshape(-1,1)
x1=datos2['CRIM'].values.reshape(-1,1)
x2=datos2['ZN'].values.reshape(-1,1)
x3=datos2['INDUS'].values.reshape(-1,1)
x4=datos2['CHAS'].values.reshape(-1,1)
x5=datos2['NOX'].values.reshape(-1,1)
x6=datos2['RM'].values.reshape(-1,1)
x7=datos2['AGE'].values.reshape(-1,1)
x8=datos2['DIS'].values.reshape(-1,1)
x9=datos2['RAD'].values.reshape(-1,1)
x10=datos2['TAX'].values.reshape(-1,1)
x11=datos2['PTRATIO'].values.reshape(-1,1)
x12=datos2['B'].values.reshape(-1,1)
x13=datos2['LSTAT'].values.reshape(-1,1)
X=np.hstack([x1,x2,x3,x4,x5,x6,x7,x8,x9,x10,x11,x12,x13])
lr=LinearRegression()
lr.fit(X,y)
print(lr.score(X,y),lr.coef_,lr.intercept_)
```

```python
import statsmodels.api as sm
ones=np.ones([len(X),1])
XS=np.hstack([ones,X])
ols= sm.OLS(y,XS)
results=ols.fit()
results.summary()
```
